# Density and Contour Plot Tutorial

https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
  
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

#print(X.shape, X)
#print(Y.shape, Y)

plt.contour(X, Y, Z, colors='black'); # b&w countour plot, dotted lines are negative values
plt.show()
plt.contour(X, Y, Z, 20, cmap='RdGy'); # 20 equally spaced lines within the data interval, Colormap is Red Gray
plt.show()
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar(); # black regions are data peaks and red regions are data valleys
plt.show()
```

```
plt.imshow(Z, 
           extent=[0,5,0,5],  # extern=[xmin,xmax,ymin,ymax]
           origin='lower',    # default for origin is upper left
           cmap='RdGy');
plt.colorbar();
plt.axis(aspect='image'); # this ensures that x and y units match
```

```
contours = plt.contour(X, Y, Z, 3, colors='black');
plt.clabel(contours, inline=True, fontsize=8);
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy', alpha=0.5); # let's the lines show through
plt.colorbar();
```
